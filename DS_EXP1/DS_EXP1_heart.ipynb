```python
df['Sex'] = df['Sex'].replace({'F':0,'M':1})
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le1 = LabelEncoder()

df['ChestPainType'] = le1.fit_transform(df['ChestPainType'])

df.head()
```

```python

label_mapping = dict(zip(le1.classes_, range(len(le1.classes_))))
print(label_mapping)
```

```python
le1.inverse_transform([0,1,2,3])
```

```python
df['ExerciseAngina'] = df['ExerciseAngina'].replace({'N': 0, 'Y': 1}).astype(int)
df.head()
```

```python
df['RestingECG'].unique()
```

```python
df['RestingECG'] = df['RestingECG'].replace({'Normal':0,'ST':1,'LVH':2})
df.head()
```

```python
df['ST_Slope'].unique()
```

```python
df['ST_Slope'] = df['ST_Slope'].replace({'Down': -1,'Flat': 0,'Up': 1})
```
